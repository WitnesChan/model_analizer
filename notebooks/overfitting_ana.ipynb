```python
import pickle 
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
trials =pickle.load(open('trial_results_v2.sav', 'rb'))
df_trials = pd.DataFrame(trials.trials)
```

```python
df_trials = pd.concat(
   [ df_trials['misc'].apply(lambda r: pd.Series(r['vals'])).applymap(lambda r: r[0]),  df_trials['result'].apply(lambda r: r['loss']).rename('loss')],
   axis = 1
)
```

```python
df_trials
```

```python
sns.pairplot(data = df_trials, kind="kde")
```

```python
sns.regplot(data = df_trials , x = 'colsample_by_tree', y ='loss')
```

```python
sns.regplot(data = df_trials , x = 'max_depth', y ='loss')
```

```python
sns.regplot(data = df_trials , x = 'n_estimators', y ='loss')
```
